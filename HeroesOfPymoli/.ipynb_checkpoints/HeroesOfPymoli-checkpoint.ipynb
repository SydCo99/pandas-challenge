```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
#Total Players 

total_unique_SN = purchase_data['SN'].nunique()
total_players = pd.DataFrame({"Total Players": [total_unique_SN]})
total_players
```

```python
# Purchasing Analysis (Total)

# Run basic calculations to obtain number of unique items, average price, etc.
unique_items = purchase_data["Item Name"].nunique()
avg_price = purchase_data["Price"].mean()
number_of_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

#format 
avg_price_format = '${0:,.2f}'.format(avg_price)
total_revenue_format = '${0:,.2f}'.format(total_revenue)

# Create a summary data frame to hold the results
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price_format],
                                   "Number of Purchases":[number_of_purchases],"Total Revenue":[total_revenue_format]})
purchasing_analysis
```

```python
# Gender Demographics

grouped_male = purchase_data.loc[purchase_data["Gender"] == "Male", :]

male_count = grouped_male["SN"].nunique()
male_percentage = male_count / total_unique_SN
male_percentage_format = '{:,.2%}'.format(male_percentage)

grouped_female = purchase_data.loc[purchase_data["Gender"] == "Female", :]

female_count = grouped_female["SN"].nunique()
female_percentage = female_count / total_unique_SN
female_percentage_format = '{:,.2%}'.format(female_percentage)

grouped_other = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]

other_count = grouped_other["SN"].nunique()
other_percentage = other_count / total_unique_SN
other_percentage_format = '{:,.2%}'.format(other_percentage)


gender_demographics = pd.DataFrame({"Gender": ["Male", "Female", "Other/Non-Disclosed"],"Total Count":[female_count,male_count,other_count] ,"Percentage of Players":[male_percentage_format, female_percentage_format,other_percentage_format]})
gender_demographics = gender_demographics.set_index(["Gender"])
gender_demographics
```

```python
# Purchasing Analysis (Gender)

# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#male 
purchase_count_male = grouped_male["Purchase ID"].count()
avg_purchase_price_male = grouped_male["Price"].mean()
total_purchase_value_male = grouped_male["Price"].sum()
avg_total_per_male = total_purchase_value_male/male_count

#female
purchase_count_female = grouped_female["Purchase ID"].count()
avg_purchase_price_female = grouped_female["Price"].mean()
total_purchase_value_female = grouped_female["Price"].sum()
avg_total_per_female = total_purchase_value_female/female_count


#other/non-disclosed
purchase_count_other = grouped_other["Purchase ID"].count()
avg_purchase_price_other = grouped_other["Price"].mean()
total_purchase_value_other = grouped_other["Price"].sum()
avg_total_per_other = total_purchase_value_other/other_count

# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame
purchasing_by_gender = pd.DataFrame({"Gender":["Male", "Female", "Other/Non-Disclosed"],
                                    "Purchase Count":[purchase_count_male,purchase_count_female,purchase_count_other],
                                     "Average Purchase Price":[avg_purchase_price_male,avg_purchase_price_female,avg_purchase_price_other],
                                     "Total Purchase Value": [total_purchase_value_male,total_purchase_value_female,total_purchase_value_other],
                                     "Avg Total Purchase per Person":[avg_total_per_male,avg_total_per_female,avg_total_per_other]})
purchasing_by_gender = purchasing_by_gender.set_index(["Gender"])

purchasing_by_gender = purchasing_by_gender.style.format({"Average Purchase Price":"${:,.2f}",
                                  "Total Purchase Value":"${:,.2f}",
                                 "Avg Total Purchase per Person":"${:,.2f}"})

purchasing_by_gender
```

```python
# Establish bins for ages
bins = [0, 9.99, 14.99, 19.99, 24.90, 29.99, 34.99, 39.99, 1000]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age"] = pd.cut(purchase_data["Age"], bins, labels=labels)

age_demographics = pd.DataFrame(purchase_data.groupby("Age").SN.nunique())

age_demographics = age_demographics.rename(columns={"SN": "Player Count"})


age_demographics["Percentage of Players"] = age_demographics["Player Count"] / sum(age_demographics["Player Count"])*100

age_demographics = age_demographics.style.format({"Percentage of Players": "{:.2f}%"})
age_demographics
```

```python
# Purchasing Analysis (Age)
# Bin the purchase_data data frame by age
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
# Create a summary data frame to hold the results
# Optional: give the displayed data cleaner formatting
# Display the summary data frame

bins = [0, 9.99, 14.99, 19.99, 24.90, 29.99, 34.99, 39.99, 1000]
labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels = labels)
# purchase_data.set_index("Age Range")
# purchase_data.groupby("Age Range")

# Purchase Count
purchase_count = purchase_data.groupby("Age Range").count()["Purchase ID"]

# Average Purchase Price
average_price = purchase_data.groupby("Age Range").mean()["Price"]
#print(average_price)
# Total Purchase Value
total_purchase_value = purchase_data.groupby("Age Range").sum()["Price"]

# Avg Total Purchase per Person
avg_purchase_per_person = total_purchase_value/purchase_data.groupby("Age Range").SN.nunique()

age_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                      "Average Purchase Price": average_price, 
                      "Total Purchase Value": total_purchase_value, 
                      "Avg Total Purchase per Person": avg_purchase_per_person})
age_analysis = age_analysis.style.format({"Average Purchase Price": "${:,.2f}",
                                         "Total Purchase Value": "${:,.2f}",
                                         "Avg Total Purchase per Person":"${:,.2f}" })

age_analysis
```

```python
# Top Spenders
# Run basic calculations to obtain the results in the table below
#Purchase Count	Average Purchase Price	Total Purchase Value

SN_groups = purchase_data.groupby(["SN"])

purchase_count = SN_groups["Purchase ID"].count()
avg_purchase_price = SN_groups["Price"].mean()
total_purchase_value = SN_groups["Price"].sum()



# Create a summary data frame to hold the results
top_spenders = pd.DataFrame({"Purchase Count": purchase_count,
                            "Average Purchase Price":avg_purchase_price,
                            "Total Purchase Value":total_purchase_value })

top_spenders.sort_values("Total Purchase Value", ascending = False).head()

```

```python
# Retrieve the Item ID, Item Name, and Item Price columns
item_data = purchase_data[["Item ID", "Item Name", "Price"]] 
item_data
# Group by Item ID and Item Name. 
grouped_items = item_data.groupby(["Item ID", "Item Name"])
#Perform calculations to obtain purchase count, average item price, and total purchase value
purchase_count = grouped_items["Item ID"].count()
avg_item_price = grouped_items["Price"].mean()
total_purchase_value = grouped_items["Price"].sum()
# Create a summary data frame to hold the results
item_summary = pd.DataFrame({"Purchase Count":purchase_count,
                            "Average Item Price":avg_item_price,
                            "Total Purchase Value": total_purchase_value })
item_summary.sort_values("Purchase Count", ascending = False).head()
```
